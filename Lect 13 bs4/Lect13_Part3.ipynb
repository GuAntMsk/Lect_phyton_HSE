```python
def get_page(p):
    url = f'http://books.toscrape.com/catalogue/page-{p}.html'
    response = requests.get(url)
    tree = BeautifulSoup(response.content, 'html.parser')
    books = tree.find_all('article', {'class' : 'product_pod'})
    
    info = []
    
    for book in books:
        info.append({'price': book.find('p', {'class': 'price_color'}).text,
                     'href': book.h3.a.get('href'),
                     'title': book.h3.a.get('title'),
                    'rating': book.p.get('class')[1]})
                     
    return info
```

```python
from tqdm import tqdm
```

```python
import time
infa = []

for p in tqdm(range(1,51)):
    try:
        infa.extend(get_page(p))
        time.sleep(1)
    except:
        print(p)
```

```python
len(infa)
```

```python
import pandas as pd

df = pd.DataFrame(infa)
print(df.shape)
df.head()
```

Напишите программу, где пользователь вводит тег, а ему выводятся все цитаты, у которых есть такой тег (например, 'truth').

```python
import requests  

url = 'https://quotes.toscrape.com/page/1/'
response = requests.get(url)
response
```

```python
def get_quotes_on_page(p):
    url = f'https://quotes.toscrape.com/page/{p}/'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    quotes = soup.find_all('div', class_ = 'quote')
    parsed_quotes = []
    for quote in quotes:
        quote_text = quote.find('span', class_ = 'text').text
        quote_author = quote.find('small', class_ = 'author').text
        tags = [tag.text for tag in quote.find_all('a', class_ = 'tag')]
        parsed_quotes.append({'quote_text' : quote_text,
                              'author' : quote_author,
                              'tags' : tags})
    return parsed_quotes
```

```python
get_quotes_on_page(1)
```

```python
import time
infa = []

for p in tqdm(range(1,10)):
    try:
        infa.extend(get_quotes_on_page(p))
        time.sleep(1)
    except:
        print(p)
```
