```python
import random
books = {}

for i in soup.find_all('div', {'class':'product'}):
    url0 = 'https://www.bgshop.ru/' + i.find_all('a')[0].get('href')
    page0 = requests.get(url0)
    soup0 = BeautifulSoup(page0.text)
    sleep(random.randint(1, 10) / random.randint(2, 30))
    
    main_info = soup0.find('div', {'id':'productMain'})
    author = main_info.find('p').text.strip()
    title = main_info.find('h1').text.strip()
    status = main_info.find('p', {'class':'item-status'}).text
    count = re.findall(r'\d+', main_info.find_all('span')[7].text)[0]
    price = main_info.find('span', class_='price_new').text
    price_clean = float(re.findall(r'\d+.\d+', price)[0].replace(',', '.'))
    
    books[title] = [author, status, price_clean, url0, count]
```

А теперь для всех страниц с книгами:

```python
from tqdm import tqdm
```

```python
books = {}

for u in tqdm(urls):
    page = requests.get(u)
    soup = BeautifulSoup(page.text)
    sleep(1)
    info = soup.find_all('div', {'class':'product'})
    for i in info:
        url0 = 'https://www.bgshop.ru/' + i.find_all('a')[0].get('href')
        page0 = requests.get(url0)
        soup0 = BeautifulSoup(page0.text)
        sleep(1)

        main_info = soup0.find('div', {'id':'productMain'})
        author = main_info.find('p').text.strip()
        title = main_info.find('h1').text.strip()
        status = main_info.find('p', {'class':'item-status'}).text
        count = re.findall(r'\d+', main_info.find_all('span')[7].text)[0]
        price = main_info.find('span', class_='price_new').text
        price_clean = float(re.findall(r'\d+.\d+', price)[0].replace(',', '.'))

        books[title] = [author, status, price_clean, url0, count]
```

```python
len(books)
```

```python
df = pd.DataFrame(books).T.reset_index()
df.columns = ['Название', 'Автор', 'Статус', 'Цена', 'Ссылка', 'Количество']
```
