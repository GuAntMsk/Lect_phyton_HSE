## Задание 3

Создайте на основе массива полученных данных таблицу water_district.xlsx, в которой будут столбцы "Административный округ", "Район", "Среднее кол-во дней отключения воды".

Создайте на основе массива полученных данных таблицу water_admarea.xlsx, в которой будут столбцы "Административный округ" и "Среднее кол-во дней отключения воды".

```python
import pandas as pd
df = pd.DataFrame(columns = ["Административный округ", "Район", "Начало", "Конец", "Среднее"])
df
```

```python
from datetime import datetime

for el in warm_water_json['features']:
    s = 0
    for period in el['properties']['attributes']['Periods']:
        if period['OutageBegin'] == 'без отключения':
            continue
        if period['OutageEnd'] == 'без отключения':
            continue
        start = datetime.strptime(period['OutageBegin'], '%d.%m.%Y')
        end = datetime.strptime(period['OutageEnd'], '%d.%m.%Y')
        s += int((end - start).days)
    s /= len(el['properties']['attributes']['Periods'])
        
    df = df.append({"Административный округ":el['properties']['attributes']['AdmArea'], 
                           "Район":el['properties']['attributes']['District'], 
                           "Начало":el['properties']['attributes']['Periods'][0]['OutageBegin'],
                           "Конец":el['properties']['attributes']['Periods'][0]['OutageEnd'],
                           "Среднее":s}, 
                          ignore_index = True)
df.to_excel('water.xlsx', index = False)
```
